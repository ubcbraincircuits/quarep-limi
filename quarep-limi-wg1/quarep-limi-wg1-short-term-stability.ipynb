```python
# Import Packages
import pandas as pd  # for reading data files into a table
import numpy as np  # for working with data
import os  # for working with operating system (e.g. file_paths)

# Required for Matplotlib Plotting
import matplotlib as mpl  # Matplotlib "manager"
import matplotlib.pyplot as plt  # Matlab style plotting
mpl.rcParams['font.size'] = 15  # increases the font size when plotting with Matplotlib (optional - can comment out)
mpl.rcParams['figure.figsize'] = (15, 5)  # increases the figure size when plotting with Matplotlib (optional - can comment out)
```

# Quarep-limi-wg1 STAB calculations
Below is the calculated STAB values for all 4 provided .csv files. 

## 405nm Laser

```python
#Importing Data for the 405nm laser
data_path = os.path.join('..','quarep-limi','quarep-limi-405nm-noblanking.csv')

df = pd.read_csv(data_path,  sep=';',  header = 12, engine='python', skipinitialspace=True)

power = df['Power (W)'].to_numpy()
time = df['Time of day (hh:mm:ss) '].to_numpy()
```

Plotting the power versus time to manually find a threshold point for the laser, thus ignoring the 'baseline' and ensuring that when finding the P_min that it is infact the P_min of the laser when it is on.

```python
fig, ax = plt.subplots(1)
#Add Data

ax.plot(time, power, 'b', label = 'laser power')

#Format Figure
ax.set_xlabel('Time')
ax.set_ylabel('W')

ax.set_title('Power vs. Time')

ax.legend()
```

The manually found threshould value was 0.00005.

```python
pdata = np.array(power)

filtered = pdata > 0.00005 

p_min = np.min(pdata[filtered])
print(f'The value of p_min is = {p_min}') 
p_max = np.max(pdata[filtered])
print(f'The value of p_max is = {p_max}') 

STAB_s = 100*(1-((p_max-p_min))/(p_max + p_min))
print(f'The STAB_short value is = {STAB_s}') 
```

The value of p_min is = 0.00020906 <br/>
The value of p_max is = 0.00026212 <br/> 
The STAB_short value is = 88.73891081964429 <br/>
